## 30digit_2addition

```python
arithmetic = "30digit_2addition"
```

```python
def modify_30digit_2addition(data_rows):
    modified_sentences = []
    equation_results = []  # 計算結果を保存するためのリスト
    equation_counter = 1

    header = data_rows[0] #header
    data = data_rows[1:] #実データ

    for row in data:
        item_id = row[0]
        context = row[1]
        question = row[2]
        correct_answer = row[3]
        target_position = row[4]
        number_of_propositions = row[5]
        condition = row[6]
        plausibility = row[7]

        words = context.split()
        modified = []

        idx = 0
        while idx < len(words):
            # ランダムに2つの30桁の数字を生成
            num1 = random.randint(100000000000000000000000000000, 999999999999999999999999999999)
            num2 = random.randint(100000000000000000000000000000, 999999999999999999999999999999)
            result = num1 + num2 

            # 計算式を生成
            equation = [f"{num1}", "+", f"{num2}", "=", f"x{equation_counter}"]

            completed = True  # この計算問題が正常に埋め込まれたかどうかを管理
            for eq_part in equation:
                if idx < len(words):
                    modified.append(words[idx])
                    idx += 1
                if idx < len(words):  # 次の単語がある場合のみ追加
                    modified.append(eq_part)
                else:
                    completed = False  # 文が終わったため計算問題が途切れた
                    break

            if completed:
                # equation_results に元データ + equation情報を入れる
                equation_results.append([
                    f"x{equation_counter}", ' '.join(equation), result,
                    item_id, context, question, correct_answer,
                    target_position, number_of_propositions, condition, plausibility
                ])
    
            equation_counter += 1
        
        # 残りの単語を追加
        modified.extend(words[idx:])

        # 変更された文を文字列に戻す
        modified_sentence = ' '.join(modified)
        modified_sentences.append(modified_sentence)

    return modified_sentences, equation_results
```

```python
modified_sentences, equation_results = modify_30digit_2addition(df)

print("---modified_sentences---")
for modified_sentence in modified_sentences[:10]:
    print(modified_sentence)
print("---equation_results---")
for equation_result in equation_results[:10]:
  print(equation_result)
```

```python
#sentencesを保存する
new_df = copy.deepcopy(df)
new_df[0].insert(2, "Sentence_with_equations")
n=0
for row in new_df[1:]:
  row.insert(2,modified_sentences[n])
  n = n+1
print("new_df")
for row in new_df[:10]:
  print(row)

new_csv_file = "sentences_" + arithmetic + "_" + memory_load + ".csv"
with open(new_csv_file, "w", encoding = "utf-8") as f:
  writer = csv.writer(f)

  for row in new_df:
    writer.writerow(row)

#equation_resultsを保存する
answers_csv_file = "equation_" + arithmetic + "_" + memory_load + ".csv"
with open(answers_csv_file, "w", encoding = "utf-8") as f:
    writer = csv.writer(f)
    writer.writerow([
        "EquationID", "Equation", "Result",
        "ItemID", "Context", "Question", "Correct_answer",
        "Target_sentence_position", "Number_of_propositions", "Condition", "Plausibility"
    ])
    for row in equation_results:
        writer.writerow(row)
```

```python
def modify_1digit_3addition(data_rows):
    modified_sentences = []
    equation_results = []  # 計算結果を保存するためのリスト
    equation_counter = 1

    header = data_rows[0] #header
    data = data_rows[1:] #実データ

    for row in data:
        item_id = row[0]
        context = row[1]
        question = row[2]
        correct_answer = row[3]
        target_position = row[4]
        number_of_propositions = row[5]
        condition = row[6]
        plausibility = row[7]

        words = context.split()
        modified = []

        idx = 0
        while idx < len(words):
            # ランダムに3つの3桁の数字を生成
            num1 = random.randint(1, 9)
            num2 = random.randint(1, 9)
            num3 = random.randint(1, 9)
            result = num1 + num2 + num3

            # 計算式を生成
            equation = [f"{num1}", "+", f"{num2}", "+", f"{num3}", "=", f"x{equation_counter}"]

            completed = True  # この計算問題が正常に埋め込まれたかどうかを管理
            for eq_part in equation:
                if idx < len(words):
                    modified.append(words[idx])
                    idx += 1
                if idx < len(words):  # 次の単語がある場合のみ追加
                    modified.append(eq_part)
                else:
                    completed = False  # 文が終わったため計算問題が途切れた
                    break

            if completed:
                # ✅ equation_results に元データ + equation情報を入れる
                equation_results.append([
                    f"x{equation_counter}", ' '.join(equation), result,
                    item_id, context, question, correct_answer,
                    target_position, number_of_propositions, condition, plausibility
                ])
    
            equation_counter += 1
        
        # 残りの単語を追加
        modified.extend(words[idx:])

        # 変更された文を文字列に戻す
        modified_sentence = ' '.join(modified)
        modified_sentences.append(modified_sentence)

    return modified_sentences, equation_results
```

```python
modified_sentences, equation_results = modify_1digit_3addition(df)

print("---modified_sentences---")
for modified_sentence in modified_sentences[:10]:
    print(modified_sentence)
print("---equation_results---")
for equation_result in equation_results[:10]:
  print(equation_result)
```

```python
#sentencesを保存する
new_df = copy.deepcopy(df)
new_df[0].insert(2, "Sentence_with_equations")
n=0
for row in new_df[1:]:
  row.insert(2,modified_sentences[n])
  n = n+1
print("new_df")
for row in new_df[:10]:
  print(row)

new_csv_file = "sentences_" + arithmetic + "_" + memory_load + ".csv"
with open(new_csv_file, "w", encoding = "utf-8") as f:
  writer = csv.writer(f)

  for row in new_df:
    writer.writerow(row)
```

```python
#equation_resultsを保存する
answers_csv_file = "equation_" + arithmetic + "_" + memory_load + ".csv"
with open(answers_csv_file, "w", encoding = "utf-8") as f:
    writer = csv.writer(f)
    writer.writerow([
        "EquationID", "Equation", "Result",
        "ItemID", "Context", "Question", "Correct_answer",
        "Target_sentence_position", "Number_of_propositions", "Condition", "Plausibility"
    ])
    for row in equation_results:
        writer.writerow(row)
```
